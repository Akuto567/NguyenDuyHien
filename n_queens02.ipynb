## The n-Queens Problem

* __Goal:__ Find an arrangement of $n$ queens on a $n \times n$ chess board so that no queen is on the same row, column or diagonal as any other queen.

* __State space:__ An arrangement of the queens on the board. We restrict the state space to arrangements where there is only a single queen per column. We represent a state as an integer vector $\mathbf{q} = \{q_1, q_2, \dots, q_n\}$, each number representing the row positions of the queens from left to right. We will call a state a "board."

* __Objective function:__ The number of pairwise conflicts (i.e., two queens in the same row/column/diagonal).
The optimization problem is to find the optimal arrangement $\mathbf{q}^*$ of $n$ queens on the board can be written as:

  > minimize: $\mathrm{conflicts}(\mathbf{q})$
  >
  > subject to: $\mathbf{q} \ \text{contains only one queen per column}$

  Note: the constraint (subject to) is enforced by the definition of the state space.

* __Local improvement move:__ Move one queen to a different row in its column.

* __Termination:__ For this problem there is always an arrangement $\mathbf{q}^*$ with $\mathrm{conflicts}(\mathbf{q}^*) = 0$, however, the local improvement moves might end up in a local minimum.

## Helper functions

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors

np.random.seed(1234)


def random_board(n):
    """Creates a random board of size n x n. Note that only a single queen is placed in each column!"""

    return(np.random.randint(0,n, size = n))

def comb2(n): return n*(n-1)//2 # this is n choose 2 equivalent to math.comb(n, 2); // is int division

def conflicts(board):
    """Calculate the number of conflicts, i.e., the objective function."""

    n = len(board)

    horizontal_cnt = [0] * n
    diagonal1_cnt = [0] * 2 * n
    diagonal2_cnt = [0] * 2 * n

    for i in range(n):
        horizontal_cnt[board[i]] += 1
        diagonal1_cnt[i + board[i]] += 1
        diagonal2_cnt[i - board[i] + n] += 1

    return sum(map(comb2, horizontal_cnt + diagonal1_cnt + diagonal2_cnt))

# decrease the font size to fit larger boards
def show_board(board, cols = ['white', 'gray'], fontsize = 48):
    """display the board"""

    n = len(board)

    # create chess board display
    display = np.zeros([n,n])
    for i in range(n):
        for j in range(n):
            if (((i+j) % 2) != 0):
                display[i,j] = 1

    cmap = colors.ListedColormap(cols)
    fig, ax = plt.subplots()
    ax.imshow(display, cmap = cmap,
              norm = colors.BoundaryNorm(range(len(cols)+1), cmap.N))
    ax.set_xticks([])
    ax.set_yticks([])

    # place queens. Note: Unicode u265B is a black queen
    for j in range(n):
        plt.text(j, board[j], u"\u265B", fontsize = fontsize,
                 horizontalalignment = 'center',
                 verticalalignment = 'center')

    print(f"Board with {conflicts(board)} conflicts.")
    plt.show()
```

## Task 5: Simulated Annealing [10 Points]

Simulated annealing is a form of stochastic hill climbing that avoid local optima by also allowing downhill moves with a probability proportional to a temperature. The temperature is decreased in every iteration following an annealing schedule. You have to experiment with the annealing schedule (Google to find guidance on this).


1. Implement simulated annealing for the n-Queens problem.
2. Create a visualization of the search process (a line chart of how the number if conflict changes as the algorithm progrsses).
3. Use this visualization for experiments with different choices for the annealing schedule and discuss what you have learned.

```python
# Code and description go here
import random
import math
import matplotlib.pyplot as plt

def random_state(n):
    return [random.randint(0, n - 1) for _ in range(n)]

def conflicts(state):
    n = len(state)
    count = 0
    for i in range(n):
        for j in range(i + 1, n):
            if state[i] == state[j] or abs(state[i] - state[j]) == abs(i - j):
                count += 1
    return count

def random_neighbor(state):
    n = len(state)
    new_state = state.copy()
    col = random.randint(0, n - 1)
    new_state[col] = random.randint(0, n - 1)
    return new_state

def simulated_annealing(n, T0=100, alpha=0.99, max_iter=10000):
    current = random_state(n)
    T = T0
    current_conflicts = conflicts(current)
    conflict_history = [current_conflicts]

    for step in range(max_iter):
        if current_conflicts == 0:
            break

        neighbor = random_neighbor(current)
        neighbor_conflicts = conflicts(neighbor)
        delta = neighbor_conflicts - current_conflicts

        # Accept if better or with probability exp(-ΔE/T)
        if delta < 0 or random.random() < math.exp(-delta / T):
            current = neighbor
            current_conflicts = neighbor_conflicts

        conflict_history.append(current_conflicts)

        # Decrease temperature (annealing schedule)
        T = T * alpha

        # Optional stop: temperature too low
        if T < 1e-5:
            break

    return current, current_conflicts, conflict_history

# --- Thực thi ---
n = 8  # 8-Queens
solution, final_conflicts, history = simulated_annealing(n, T0=100, alpha=0.98, max_iter=5000)

print("Final solution:", solution)
print("Conflicts:", final_conflicts)

# --- Vẽ biểu đồ ---
plt.plot(history)
plt.title(f"Simulated Annealing for {n}-Queens")
plt.xlabel("Iteration")
plt.ylabel("Number of Conflicts")
plt.grid(True)
plt.show()
```

```python
# Code
import random
import math
import time
import numpy as np

# --- Các hàm cơ bản cho N-Queens ---
def random_state(n):
    return [random.randint(0, n - 1) for _ in range(n)]

def conflicts(state):
    n = len(state)
    count = 0
    for i in range(n):
        for j in range(i + 1, n):
            if state[i] == state[j] or abs(state[i] - state[j]) == abs(i - j):
                count += 1
    return count

def neighbors(state):
    n = len(state)
    result = []
    for c in range(n):
        for r in range(n):
            if state[c] != r:
                new_state = state.copy()
                new_state[c] = r
                result.append(new_state)
    return result

# --- Steepest Ascent Hill Climbing ---
def steepest_ascent_hc(n):
    current = random_state(n)
    current_conf = conflicts(current)
    while True:
        neighs = neighbors(current)
        best = min(neighs, key=conflicts)
        best_conf = conflicts(best)
        if best_conf >= current_conf:
            return current, current_conf
        current, current_conf = best, best_conf

# --- Stochastic Hill Climbing 1: chọn ngẫu nhiên neighbor tốt hơn ---
def stochastic_hc1(n):
    current = random_state(n)
    current_conf = conflicts(current)
    while True:
        neighs = neighbors(current)
        better = [s for s in neighs if conflicts(s) < current_conf]
        if not better:
            return current, current_conf
        current = random.choice(better)
        current_conf = conflicts(current)

# --- Stochastic Hill Climbing 2: chọn ngẫu nhiên 1 neighbor và chấp nhận nếu tốt hơn ---
def stochastic_hc2(n, max_iter=1000):
    current = random_state(n)
    current_conf = conflicts(current)
    for _ in range(max_iter):
        new = random_neighbor(current)
        new_conf = conflicts(new)
        if new_conf < current_conf:
            current, current_conf = new, new_conf
    return current, current_conf

def random_neighbor(state):
    n = len(state)
    new_state = state.copy()
    col = random.randint(0, n - 1)
    new_state[col] = random.randint(0, n - 1)
    return new_state

# --- Simulated Annealing ---
def simulated_annealing(n, T0=100, alpha=0.98, max_iter=10000):
    current = random_state(n)
    T = T0
    current_conf = conflicts(current)
    for step in range(max_iter):
        if current_conf == 0:
            break
        neighbor = random_neighbor(current)
        neighbor_conf = conflicts(neighbor)
        delta = neighbor_conf - current_conf
        if delta < 0 or random.random() < math.exp(-delta / T):
            current, current_conf = neighbor, neighbor_conf
        T = T * alpha
        if T < 1e-5:
            break
    return current, current_conf

# --- Đo thời gian và thống kê ---
def test_algorithm(algorithm, n, runs=100):
    times = []
    confs = []
    success = 0
    for _ in range(runs):
        start = time.time()
        sol, conf = algorithm(n)
        times.append(time.time() - start)
        confs.append(conf)
        if conf == 0:
            success += 1
    avg_time = np.mean(times)
    avg_conf = np.mean(confs)
    success_rate = success / runs * 100
    return avg_time, avg_conf, success_rate

# --- Chạy và thu kết quả ---
algorithms = {
    "Steepest asc. HC": steepest_ascent_hc,
    "Stochastic HC 1": stochastic_hc1,
    "Stochastic HC 2": stochastic_hc2,
    "Simulated Annealing": simulated_annealing
}

board_sizes = [4, 8]

print("| Algorithm | Board size | Avg. Run time | Avg. number of conflicts | % Optimal |")
print("|------------|-------------|---------------|---------------------------|-----------|")

for name, func in algorithms.items():
    for n in board_sizes:
        avg_time, avg_conf, success_rate = test_algorithm(func, n)
        print(f"| {name:18} | {n:^11} | {avg_time:13.4f} | {avg_conf:25.2f} | {success_rate:9.1f} |")
```
